# 1. What is need for Hypothesis Testing ?  

Need for Hypothesis Testing

Hypothesis testing is needed because in real-world problems, you almost never have complete data about the entire population—you only have a sample. Making decisions directly from that sample without validation is risky and often wrong.

So hypothesis testing gives you a structured, mathematical way to decide whether what you’re seeing is real or just random chance.

Break it down properly:

To make decisions under uncertainty

Example: You collected crop yield data from a few farms.

Question: Is a new fertilizer actually improving yield, or is this just random variation?

Hypothesis testing helps you decide with statistical confidence.

To validate assumptions (not blindly trust them)

Instead of saying “I think the average is higher,” you formally test:

Null hypothesis (H₀): No effect / no difference

Alternative hypothesis (H₁): There is an effect

To avoid wrong conclusions from random noise

Small samples can be misleading.

Hypothesis testing reduces the risk of:

False positives (Type I error)

False negatives (Type II error)

To support data-driven decisions

Used in:

Business (A/B testing)

Agriculture (yield comparison)

Medicine (drug effectiveness)

Tech (model evaluation)

To quantify evidence (not guess)

Gives you tools like:

p-value

significance level (α)

So instead of saying “it looks better,” you say:
→ “There is statistically significant evidence at α = 0.05”

# 3.What is a student t distribution?

Student’s t-Distribution

The Student’s t-distribution is a probability distribution used in hypothesis testing when:

The sample size is small (n < 30)

The population standard deviation is unknown

Why it exists (this is where most students fail)

In real life, you almost never know the population standard deviation. So instead of using a z-test, you estimate it from the sample—and that introduces extra uncertainty.

The t-distribution adjusts for this uncertainty by having heavier tails than the normal distribution.

Key Characteristics

Bell-shaped and symmetric (like normal distribution)

Heavier tails

More spread → accounts for higher uncertainty

Depends on Degrees of Freedom (df)

df = n − 1

Smaller sample → wider curve

As sample size increases → it approaches normal distribution

# 4.What is level of significance ? What role t plays in hypothesis Testing

Level of Significance (α)

The level of significance (α) is the maximum probability of making a wrong decision by rejecting a true null hypothesis.

In simple terms:
→ It is the risk you are willing to take of being wrong.

Key Points

Common values: 0.05 (5%), 0.01 (1%)

It represents Type I Error

Set before performing the test

Interpretation

α = 0.05 → You accept a 5% chance of false positive

Lower α → stricter test (harder to reject H₀)
