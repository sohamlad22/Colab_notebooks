###**Boxplot Analysis:**

A boxplot is a graphical representation used to visualize the distribution of numerical data. It displays the following statistical measures:

* Minimum value
* First Quartile (Q1)
* Median (Q2)
* Third Quartile (Q3)
* Maximum value
* Outliers (if present)

Boxplots help in identifying:

* Spread of the data
* Skewness of distribution
* Presence of extreme values (outliers)

In this analysis, boxplot is used to detect outliers in the student dataset and understand the distribution of numerical attributes.

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

The code first creates a figure of size 8×5 inches using `plt.figure(figsize=(8,5))` so that the boxplot is clearly visible.

Then `sns.boxplot(data=df)` generates boxplots for all numerical columns in the dataframe, helping visualize the distribution and detect outliers.

The `plt.xticks(rotation=90)` function rotates the x-axis labels by 90 degrees to prevent overlapping and improve readability of column names. Finally, `plt.show()` is used to display the boxplot on the screen.


###**Scatter Plot (Multivariate Analysis):**

Multivariate analysis is used to study the relationship between two or more variables.
A scatter plot is used to visualize how one numerical variable changes with respect to another numerical variable.

Each point in the scatter plot represents one observation from the dataset.
This visualization helps to:

* Identify trends
* Detect relationships between variables
* Observe clusters
* Identify outliers

In this notebook, a scatter plot is created using two numerical columns from the student dataset to analyze their relationship.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df.select_dtypes(include=np.number).columns[0],
                y=df.select_dtypes(include=np.number).columns[1],
                data=df)
plt.show()
```

This code creates a scatter plot to show the relationship between two numerical columns in the dataset. The figure size is set using `plt.figure().` `The sns.scatterplot()` function selects the first two numerical columns for the x and y axes from the dataframe. Finally, `plt.show()` displays the plot.

###**Heatmap (Correlation Visualization):**

A heatmap is a graphical representation used to visualize the correlation matrix of numerical variables.
It uses different colors to represent the strength of relationships between variables.

* Darker colors indicate stronger correlation
* Lighter colors indicate weaker correlation
* Positive values show direct relationship
* Negative values show inverse relationship

Heatmaps help in identifying which student attributes are strongly related to each other. This is useful for feature selection before applying machine learning models.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

This code generates a heatmap to visualize the correlation between numerical variables in the dataset. The `plt.figure(figsize=(8,6))` function sets the size of the plot. The `df.corr(numeric_only=True)` computes the correlation matrix for numeric columns. The `sns.heatmap()` function displays the matrix using colors, and annot=True shows the correlation values inside each cell. Finally, `plt.show()` is used to display the heatmap.
