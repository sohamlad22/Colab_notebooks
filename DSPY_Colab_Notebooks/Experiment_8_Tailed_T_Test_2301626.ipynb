**1.What is 2 tailed and  tailed T Test ?**

**Ans:**

A t-test is a statistical test used to determine whether there is a significant difference between sample data and a population parameter (usually the mean) when the population standard deviation is unknown.

A one-tailed t-test is used when the alternative hypothesis specifies a direction of the effect. It tests whether the sample mean is either significantly greater than or significantly less than the population mean, but not both. The critical region lies entirely in one tail of the probability distribution.

A two-tailed t-test is used when the alternative hypothesis does not specify a direction. It tests whether the sample mean is significantly different (either greater than or less than) from the population mean. The critical region is divided equally between both tails of the distribution.

Thus, a one-tailed test checks for a difference in a specific direction, while a two-tailed test checks for any significant difference regardless of direction.

**2.Explain what is Type I and Type II errors with an example.**

**Ans:**

**Type I and Type II Errors**

In hypothesis testing, Type I and Type II errors are the two types of errors that can occur while making decisions about a hypothesis.

A Type I error occurs when the null hypothesis (H₀) is rejected even though it is true. This means we conclude that there is an effect or difference when in reality there is none. It is also called a false positive and is represented by the level of significance (α).

A Type II error occurs when the null hypothesis (H₀) is not rejected even though it is false. This means we fail to detect an actual effect or difference. It is also called a false negative and is represented by β.

**Example:**

Consider testing whether a new medicine is effective.

Null hypothesis (H₀): The medicine has no effect.
Alternative hypothesis (H₁): The medicine is effective.

Type I error: Concluding that the medicine is effective when it actually has no effect.
Type II error: Concluding that the medicine is not effective when it actually works.

Thus, Type I error is rejecting a true hypothesis, while Type II error is failing to reject a false hypothesis.


**3.What is power of statistical test?**

**Ans:**

**Power of a Statistical Test**

The power of a statistical test is the probability of correctly rejecting the null hypothesis (H₀) when it is actually false. In other words, it measures the ability of a test to detect a true effect or difference when it exists.

Power is denoted as (1 − β), where β is the probability of a Type II error. A higher power means a lower chance of making a Type II error and a greater likelihood of identifying a significant result.

The power of a test depends on several factors such as sample size, level of significance (α), variability in the data, and the actual effect size. Increasing the sample size or effect size generally increases the power of the test.

Thus, the power of a statistical test indicates how effectively a test can identify true results.


**4.Explain input and out put characteristics for  stats.ttest_rel function in your own words.**

**Ans:**

**Input and Output Characteristics of stats.ttest_rel Function**

The *stats.ttest_rel*  function is used to perform a paired (dependent) t-test. It compares two related samples to determine whether their means are significantly different.

**Input Characteristics:**

The function takes two sets of data as input, usually in the form of arrays, lists, or pandas series. These two datasets must be of the same size because each value in one sample is paired with a corresponding value in the other sample. The data typically represents measurements taken from the same group at two different times or under two different conditions. Optional parameters like axis and handling of missing values can also be provided.

**Output Characteristics:**

The function returns two main values: the t-statistic and the p-value. The t-statistic indicates the magnitude and direction of the difference between the paired samples. The p-value helps in deciding whether the difference is statistically significant. If the p-value is less than the chosen significance level (such as 0.05), the null hypothesis is rejected; otherwise, it is not rejected.

Thus, stats.ttest_rel takes paired data as input and provides statistical evidence (t-statistic and p-value) to determine whether there is a significant difference between the two related samples.

**Function for Identifying Power of a Statistical Test**

In Python, the statsmodels module provides functions to calculate the power of a statistical test. One commonly used function is ttest_power() (or through TTestPower class in statsmodels.stats.power), which is used to compute the power of a t-test.

**Output Range:**

The output of the power function lies between 0 and 1.

* A value close to 0 indicates very low power, meaning the test is unlikely to detect a true effect.

* A value close to 1 indicates high power, meaning the test is very likely to detect a true effect.

**Significance:**
The power of a test represents the probability of correctly rejecting the null hypothesis when it is false. It is equal to (1 − β), where β is the probability of a Type II error. A higher power is desirable because it reduces the chances of missing a real effect. In practice, a power of 0.8 (80%) or higher is generally considered acceptable.

Thus, power functions in the stats module help in evaluating how effective a statistical test is in detecting meaningful differences or effects.
