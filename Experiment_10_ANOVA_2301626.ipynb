##1. Explain input and output characteristics for f_oneway function from stats library

Ans:

**1. Definition**
`f_oneway()` is a function from the **SciPy stats library** used to perform a **one-way ANOVA test**.  
It determines whether the **means of two or more independent groups are significantly different**.

**2. Input Characteristics**
**Syntax:** stats.f_oneway(sample1, sample2, sample3, ...)

**Input Requirements:**
1. **Multiple Groups** – Two or more groups (e.g., race, class, region)
2. **Numeric Data** – Each group must contain numeric values (age, marks, income)
3. **Independent Samples** – Data in one group should not affect others
4. **Flexible Data Types** – List, NumPy array, or Pandas Series
5. **Unequal Sample Sizes Allowed** – Groups can have different lengths

**Example Input:**

```python
group1 = [23, 25, 27, 30]
group2 = [22, 24, 29, 31]
group3 = [20, 21, 23, 26]

import scipy.stats as stats
stats.f_oneway(group1, group2, group3)
```


**3. Output Characteristics**
**Return Format:** F_onewayResult(statistic=F_value, pvalue=p_value)

**Components:**
1. **F-statistic** – Ratio of variance between groups / variance within groups
   - Large F → Means differ
   - Small F → Means are similar
2. **p-value** – Probability differences occurred by chance
   - p < 0.05 → Significant (Reject H₀)
   - p ≥ 0.05 → Not significant

**4. Example Output**
F_onewayResult(statistic=1.77, pvalue=0.13)

**Explanation:**  
- F = 1.77 → Low variation between groups  
- p = 0.13 (> 0.05) → Not statistically significant  
**Conclusion:** No significant difference between group means



**Summary**

| Aspect     | Description                                |
|------------|--------------------------------------------|
| Input      | 2+ independent numeric groups              |
| Output     | F-statistic and p-value                    |
| Purpose    | Compare group means                        |
| Limitation | Cannot identify specific group differences |

##2. Explain results of sm.stats.anova_lm fucntion.

Ans:
**Definition**

sm.stats.anova_lm() is a function from the statsmodels that performs ANOVA (Analysis of Variance) using a fitted linear model.

* It is used to determine whether the means of a numeric variable differ across categories of a factor and provides detailed table-based results including sum of squares, F-statistic, and p-value.

**Input Characteristics**
 * Syntax
   
   sm.stats.anova_lm(model, typ=2)
* model → A fitted linear model created with ols()
* typ → Type of ANOVA (1, 2, or 3); most common: typ=2

**Input Requirements**

1) fitted Model
* Must be created using statsmodels.formula.api.ols()
    
    model = ols('age ~ race', data=voter_frame).fit()
2) Factor Variables
* Independent categorical variable(s) (e.g., race)
* Dependent numeric variable (e.g., age)
3) Optional Parameter typ
* typ=2 → Most common, provides Type II sum of squares

**Example Input**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

# Create a sample voter_frame DataFrame to demonstrate the ANOVA
voter_frame = pd.DataFrame({
    'age': [25, 30, 35, 40, 28, 32, 38, 42, 26, 31, 36, 41],
    'race': ['Asian', 'Asian', 'White', 'White', 'Black', 'Black', 'Asian', 'Asian', 'White', 'White', 'Black', 'Black']
})

# Fit model
model = ols('age ~ race', data=voter_frame).fit()

# Perform ANOVA
anova_result = sm.stats.anova_lm(model, typ=2)
print(anova_result)
```

**Output Components Explained**

**1. sum_sq (Sum of Squares)**
* Variation explained by each source
* Between-group variation → race
* Within-group variation → Residual

**2.df (Degrees of Freedom)**
* Number of independent pieces of information for each source
* race df = number of groups - 1
* Residual df = total observations - number of groups

**3.F (F-statistic)**
* Ratio of variance explained by factor / residual variance
* High F → Strong evidence that groups differ

**4.PR(>F) (p-value)**
* Probability that differences occurred by chance
* p < 0.05 → Reject null hypothesis (means differ)
* p ≥ 0.05 → Fail to reject null hypothesis

**Explanation of Example**

* Sum of Squares
  * race = 199.369 → Variation in ages explained by race
  * Residual = 27948.102 → Variation within groups
* F-statistic = 1.774
  * Small → Low variation explained by race compared to residual
* p-value = 0.1317 (> 0.05)
  * Not statistically significant

##3. What role ANOVA plays in data analytics?

Ans:
**Role of ANOVA in Data Analytics**

**1. Definition**

ANOVA (Analysis of Variance) is a statistical method used in data analytics to compare the means of two or more groups to see if there are statistically significant differences among them.

It generalizes the t-test (which compares only two groups) to three or more groups.

**2. Primary Role in Data Analytics**

**Identify Group Differences**  
- Determines whether a categorical factor (e.g., region, age group, race) affects a numeric variable (e.g., sales, test scores, revenue).

**Test Hypotheses**  
- Null hypothesis (H0) → All group means are equal  
- Alternative hypothesis (H1) → At least one group mean is different

**Guide Decision Making**  
- Helps businesses, researchers, and analysts identify patterns and differences across categories.  
- Example: Which marketing campaign performs better across regions?

**Foundation for Post-hoc Analysis**  
- If ANOVA is significant, analysts perform post-hoc tests (e.g., Tukey HSD) to identify which specific groups differ.


**3. Why ANOVA is Important in Data Analytics**

| Role                        | Explanation |
|------------------------------|------------|
| Handle Multiple Groups       | Compares means across 3+ groups without inflating Type I error |
| Identify Factors Affecting Outcomes | Helps understand which factors influence the target variable |
| Reduce False Positives       | Avoids multiple pairwise t-tests which increase error probability |
| Quantitative Insights        | Provides F-statistic and p-value for rigorous decision-making |
| Visualization                | Can be combined with plots (boxplots, bar charts) for clear insights |



**4. Example in Real-World Data Analytics**

**Retail Sales Analysis**  
- Compare average sales across different store locations or marketing campaigns  
- ANOVA can tell whether location or campaign type significantly affects sales.

**Education Analytics**  
- Compare average test scores among students from different teaching methods or classrooms.

**Healthcare Analytics**  
- Compare average recovery times across multiple treatment methods or hospitals.




##4. When to use one way ANOVA? Explain with real world case.

Ans:


**1. Definition**

**One-Way ANOVA** is used to compare the means of **three or more independent groups** to determine if at least one group mean is significantly different from the others.

It is called “one-way” because there is **only one independent (categorical) variable** or factor.



**2. When to Use One-Way ANOVA**

Use **One-Way ANOVA** when:

1. **You have one categorical factor** with 3+ levels (groups)  
   - Example: Type of treatment, region, department, or marketing campaign.

2. **You have one numeric dependent variable**  
   - Example: Sales, age, test scores, recovery time, income.

3. **You want to test for differences among group means** instead of comparing each pair separately (to avoid inflated Type I error).

4. **Groups are independent**  
   - Observations in one group do not affect another.

5. **Data roughly meets ANOVA assumptions**  
   - Normal distribution within groups  
   - Homogeneity of variances (similar spread across groups)



**3. Real-World Case Example**

**Scenario:** A company wants to compare the **effectiveness of three different training programs** on employee performance scores.

- **Independent variable (factor):** Training program (Program A, B, C)  
- **Dependent variable:** Employee performance score (numeric)

**Step 1 – Question:**  
- Does the **average performance** differ among the three training programs?

**Step 2 – Why One-Way ANOVA:**  
- Only one factor (training program)  
- More than 2 groups (A, B, C)  
- Numeric outcome (performance score)  
- Avoids running multiple t-tests, reducing false positives

**Step 3 – Interpretation:**  
- **F-statistic and p-value** tell if there is a **significant difference among the training programs**.  
- If p < 0.05 → At least one program leads to a different mean score.  
- Post-hoc test (e.g., Tukey HSD) identifies **which programs differ**.

##5. When to use 2 way ANOVA ? Explain with real world case.

Ans:

**1. Definition**

Two-Way ANOVA is used to determine the effect of **two independent (categorical) variables** on a single numeric dependent variable.

It also tests whether there is an **interaction effect** between the two factors — meaning the effect of one factor may depend on the level of the other factor.



**2. When to Use Two-Way ANOVA**

Use Two-Way ANOVA when:

1. **You have two categorical independent variables (factors)**  
   - Example: Gender (Male/Female) and Teaching Method (Method A/B)

2. **You have one numeric dependent variable**  
   - Example: Test scores, sales, revenue, recovery time

3. **You want to examine:**  
   - The main effect of each factor individually  
   - The interaction effect between the two factors

4. **Groups are independent**  
   - Observations in one group do not affect another

5. **Data roughly meets ANOVA assumptions**  
   - Normality within groups  
   - Homogeneity of variances  
   - Independent observations



**3. Real-World Case Example**

**Scenario:** A school wants to compare student test scores based on:

- **Factor 1:** Teaching Method (Method A, Method B)  
- **Factor 2:** Gender (Male, Female)

**Step 1 – Question:**  
- Does test score differ based on **teaching method**, **gender**, or their **interaction**?

**Step 2 – Why Two-Way ANOVA:**  
- Two factors (Teaching Method, Gender)  
- Numeric outcome (Test Score)  
- Allows testing **both main effects and interaction effect** simultaneously  
- More efficient than multiple one-way ANOVAs

**Step 3 – Interpretation:**  
- **Main Effect 1:** Does teaching method affect scores regardless of gender?  
- **Main Effect 2:** Does gender affect scores regardless of teaching method?  
- **Interaction Effect:** Does the effect of teaching method differ for males vs females?
